This script is to scrape relevant information of posted "data scientist" positions in US from [glassdoor](https://www.glassdoor.ca/Job/us-data-scientist-jobs-SRCH_IL.0,2_IN1_KO3,17.htm) using selenium. Credits to [Ã–mer Sakarya](https://github.com/arapfaik/scraping-glassdoor-selenium) for providing the original script and tutorial from [Ken Jee](https://www.youtube.com/c/KenJee1)

Pre-requisites include installations of Python, Jupyter notebook, Selenium library and [chromedriver](https://sites.google.com/a/chromium.org/chromedriver/downloads). The installation of selenium library on a windows machine is a bit tricky. I first followed instructiosn from [Jonanthon Soma](http://jonathansoma.com/lede/foundations-2018/classes/selenium/selenium-windows-install/) using `pip` which works on local IDE but not in Juputer notebook, I figured out later that installation using `pip` caused a difference in versions and as Jupyter was launched from Anaconda, it was not able to get the correct version. By reinstalling Selenium using `conda` will solve this problem. Remember use Anaconda prompt.
